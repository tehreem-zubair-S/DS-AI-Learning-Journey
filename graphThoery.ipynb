```python
import numpy as np
import matplotlib.pyplot as plt
import networkx as nx

# Create a sample graph
G = nx.Graph()
edges = [(0,1), (0,2), (1,2), (1,3), (2,3), (3,4)]
G.add_edges_from(edges)

# Adjacency Matrix
A = nx.to_numpy_array(G, dtype=int)

print("Adjacency Matrix A:")
print(A)

# Number of steps
r = 3

# Compute A^r
Ar = np.linalg.matrix_power(A, r)

print(f"\nA^{r} (Number of paths of length {r}):")
print(Ar)

# Reachability output
n = A.shape[0]
print(f"\nReachability in exactly {r} steps:")
for i in range(n):
    for j in range(n):
        if Ar[i][j] > 0:
            print(f"Node {i} can reach Node {j} in {r} steps")

# Graph visualization
plt.figure()
nx.draw(G, with_labels=True, node_color='lightblue', node_size=1500)
plt.title("Graph Representation")
plt.show()

# Heatmap of A^r
plt.figure()
plt.imshow(Ar)
plt.colorbar()
plt.title(f"Heatmap of A^{r} (Path Counts)")
plt.show()
```
